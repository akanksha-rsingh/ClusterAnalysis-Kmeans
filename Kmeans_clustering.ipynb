```
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings("ignore")

from google.colab import drive
drive.mount('/gdrive')
#Change current working directory to gdrive
%cd /gdrive
```

```
#Read files
trainfile = r'/gdrive/My Drive/CIS508/Assignment 4/Sales_Transactions_Dataset_Weekly.csv'
trainData = pd.read_csv(trainfile) #creates a dataframe
# String value hence removed from the train data set.
X_train = trainData
X_train = trainData.drop(columns="Product_Code", axis = 1)

```

```


print(X_train.shape)
print(X_train.head())

```

```
print(X_train.shape)
```

```
#K-Means Clustering ========================================================================
#Default Mode - K=8
kmeans = KMeans()
kmeans.fit(X_train)
kmeans.labels_
kmeans.cluster_centers_
#Print Silhouette measure
print(silhouette_score(X_train, kmeans.labels_))
```

```
Kmeans = KMeans(n_clusters=3, n_init=5, init= 'k-means++', algorithm='auto', max_iter= 200)
kmeans.fit(X_train)
kmeans.labels_
kmeans.cluster_centers_
#Print Silhouette measure
print(silhouette_score(X_train, kmeans.labels_))
```

```
#Add Cluster Number to each datapoint + save file
df_kmeans=X_train
kmeans_predict_train=kmeans.predict(X_train)

df_kmeans['Cluster Number with HPT']=kmeans_predict_train
print(df_kmeans.shape)
df_kmeans.head()
df_kmeans.dtypes

export_csv = df_kmeans.to_csv(r'/gdrive/My Drive/CIS508/Assignment 4/Sales_Transactions_Dataset_Weekly.csv')
```
